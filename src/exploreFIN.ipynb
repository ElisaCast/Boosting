```python
import pandas as pd

train_data = pd.read_csv("C:/Users/elisa/OneDrive/Escritorio/Data Sciencie/ML-RandomForest/data/processed/clean_train.csv")
test_data = pd.read_csv("C:/Users/elisa/OneDrive/Escritorio/Data Sciencie/ML-RandomForest/data/processed/clean_test.csv")

train_data.head()
```

```python
X_train = train_data.drop(["Outcome"], axis = 1)
y_train = train_data["Outcome"]
X_test = test_data.drop(["Outcome"], axis = 1)
y_test = test_data["Outcome"]
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(n_estimators = 100, learning_rate = 0.01, loss = 'deviance', subsample = 1.0, max_depth = 3, random_state = 42)
# Número de árboles en el ensamble, Tasa de aprendiz, Función de pérdida (puede ser 'deviance' o 'exponential', entre otras),
# Proporción de muestras utilizadas para ajustar cada árbol (1.0 utiliza todas las muestras),Profundidad máxima de los árboles,
# Semilla para reproducibilidad

# Entrenamiento del modelo
model.fit(X_train, y_train)

# Predicciones
y_pred = model.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

Conclusiones de las tres graficas: Arbol de decision, Random Forest y Boosting 

XGBoost vs. Random Forest:

XGBoost generalmente ofrece un rendimiento robusto y a menudo supera a Random Forest debido a su capacidad para manejar relaciones no lineales y su enfoque de optimización de funciones.
Random Forest es un modelo fuerte que puede proporcionar un buen rendimiento en muchos casos, pero puede ser superado por XGBoost en ciertas situaciones.

Random Forest vs. Decision Tree:

Random Forest utiliza múltiples árboles para realizar predicciones, lo que tiende a ser más robusto y generalizable en comparación con un único árbol de decisión.
Un único árbol de decisión puede ser propenso a sobreajuste y menos capaz de capturar patrones complejos en los datos.

XGBoost vs. Decision Tree:

XGBoost es más sofisticado y potente que un único árbol de decisión, ya que utiliza un conjunto de árboles débiles y técnicas de regularización para mejorar la generalización.
Un único árbol de decisión puede ser limitado en su capacidad para manejar relaciones complejas y puede estar más inclinado hacia el sobreajuste.

Como conclusión final y según los datos obtenidos podemos predecir que XGBoost es un modelo mejorado y con mejores resultados a la hora del entrenamiento.
